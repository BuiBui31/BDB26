```python
#Imports
import pandas as pd
from pathlib import Path
import os
import numpy as np
```

```python
#Read in Data
projectRoot = Path.cwd()
dfPlays = pd.read_csv(f'{projectRoot}/cleanedData/dfPlaysWithoutPen.csv')
len(dfPlays)
```

```python
# Keep plays that have pre-snap WP ≤ 0.95
df = dfPlays.copy()
wp_max = pd.concat([
    pd.to_numeric(df['pre_snap_home_team_win_probability'], errors='coerce'),
    pd.to_numeric(df['pre_snap_visitor_team_win_probability'], errors='coerce')
], axis=1).max(axis=1)
df =  df[(wp_max <= 0.95)]
len(df)
```

```python
#No Sacks or Scrambles
dropBack = df['pass_result'].astype(str).str.upper()
dfPlays = df[(~dropBack.isin(['S','R']))].copy()
len(dfPlays)
```

```python
#are NOT scrambles/aborts and
dropBack = df['dropback_type'].astype(str).str.upper()
dfPlays = df[(~dropBack.isin(['SCRAMBLE','ABORT']))].copy()
len(dfPlays)
```

```python
#No penalties
dfPlays = dfPlays[(dfPlays['penalty_yards'].isna())].copy()
len(dfPlays)
```

```python
# only plays over los
pl = pd.to_numeric(df['pass_length'], errors='coerce')
dfPlays = dfPlays[pl > 0].copy()
len(dfPlays)
```
